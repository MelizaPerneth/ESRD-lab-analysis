```python
df = pd.read_csv('CausesofDeath.csv')
df.head()
```

```python
# Columnas que se deben mantener intactas
columnas_identificadoras = ['Entity', 'Code', 'Year', 'Number of executions (Amnesty International)']

# Realizar melt, excluyendo las columnas identificadoras
df_vertical = df.melt(id_vars=columnas_identificadoras,
                      var_name='Causes_of_Death',
                      value_name='Deaths')

# Eliminar texto innecesario de las causas de muerte
df_vertical['Causes_of_Death'] = df_vertical['Causes_of_Death'].str.replace(
    r'Deaths - | - Sex: Both - Age: All Ages \(Number\)', '', regex=True)

# Mostrar el dataframe transformado
print("\nDataframe transformado:")
print(df_vertical.head())

# Guardar el resultado a un nuevo archivo CSV
df_vertical.to_csv('dataset_transformado.csv', index=False)
print("\nEl dataframe transformado se ha guardado como 'dataset_transformado.csv'.")
```

**Eliminar columnas irrelevantes**

```python
data = data.drop(columns=['Code', 'Number of executions (Amnesty International)'])
```

```python
data.head()
```

```python
# Manejar valores nulos en la columna 'Deaths'
data['Deaths'] = data['Deaths'].fillna(0)
```

```python
data.isnull().sum()
```

```python
# Renombrar la columna 'Entity' a 'Area'
data = data.rename(columns={'Entity': 'Area'})
```

```python
data.head()
```

```python
# Filtrar los aos entre 2015 y 2019
filtered_data = data[(data['Year'] >= 2015) & (data['Year'] <= 2019)]

# Agrupar por causas de muerte y sumar las muertes
grouped_data = filtered_data.groupby('Causes_of_Death')['Deaths'].sum().reset_index()

# Ordenar por nmero de muertes en orden descendente
grouped_data = grouped_data.sort_values(by='Deaths', ascending=False)

# Calcular el porcentaje y porcentaje acumulado
grouped_data['Percentage'] = (grouped_data['Deaths'] / grouped_data['Deaths'].sum()) * 100
grouped_data['Cumulative_Percentage'] = grouped_data['Percentage'].cumsum()

print(grouped_data.head())
```

```python
# Filtrar los aos entre 2015 y 2019
filtered_data = data[(data['Year'] >= 2015) & (data['Year'] <= 2019)]

# Agrupar por rea y Causa de Muerte y sumar las muertes
grouped_data = filtered_data.groupby(['Area', 'Causes_of_Death'])['Deaths'].sum().reset_index()

# Funcin para graficar Pareto por rea
def plot_pareto_by_area_interactive(area_name, show_top_5=False):
    # Filtrar datos por el rea seleccionada
    area_data = grouped_data[grouped_data['Area'] == area_name]

    # Agrupar nuevamente solo por las causas de muerte
    area_grouped = area_data.groupby('Causes_of_Death')['Deaths'].sum().reset_index()

    # Ordenar los datos en orden descendente por nmero de muertes
    area_grouped = area_grouped.sort_values(by='Deaths', ascending=False)

    # Filtrar el top 5 si est habilitado
    if show_top_5:
        area_grouped = area_grouped.head(5)

    # Calcular el porcentaje y porcentaje acumulado
    area_grouped['Percentage'] = (area_grouped['Deaths'] / area_grouped['Deaths'].sum()) * 100
    area_grouped['Cumulative_Percentage'] = area_grouped['Percentage'].cumsum()

    # Crear el grfico de Pareto
    sns.set(style="whitegrid")
    fig, ax1 = plt.subplots(figsize=(12, 8))

    # Histograma (barras) para muertes
    ax1.bar(area_grouped['Causes_of_Death'], area_grouped['Deaths'], color='skyblue', alpha=0.7, label='Deaths')
    ax1.set_xlabel('Causes of Death', fontsize=12)
    ax1.set_ylabel('Number of Deaths', fontsize=12)
    ax1.tick_params(axis='x', rotation=90)
    ax1.set_title(f'Pareto Chart: Causes of Death in {area_name} (2015-2019)', fontsize=14)

    # Lnea del porcentaje acumulado
    ax2 = ax1.twinx()
    ax2.plot(area_grouped['Causes_of_Death'], area_grouped['Cumulative_Percentage'], color='red', marker='o', label='Cumulative Percentage')
    ax2.set_ylabel('Cumulative Percentage (%)', fontsize=12)
    ax2.set_ylim(0, 110)

    # Lnea gua al 80% (solo si no se muestra el top 5)
    if not show_top_5:
        ax2.axhline(y=80, color='green', linestyle='--', linewidth=1, label='80% Line')

    # Aadir leyendas
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    # Ajustar diseo
    plt.tight_layout()
    plt.show(
)

# Importar la funcin interact y widgets de ipywidgets
from ipywidgets import interact, widgets # Importing interact and widgets


# Crear un widget interactivo para seleccionar el rea y la opcin de top 5
areas = grouped_data['Area'].unique()
interact(
    plot_pareto_by_area_interactive,
    area_name=widgets.Dropdown(options=areas, description='Select Area:'),
    show_top_5=widgets.Checkbox(value=False, description='Show Top 5')
)
```
